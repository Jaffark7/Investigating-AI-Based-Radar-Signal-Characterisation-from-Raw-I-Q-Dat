# 01 — Explore the Data
Open the dataset and see what's inside.

```python
import sys
sys.path.append("../src")

import numpy as np
import h5py
from radar.data import load_radchar

DATA_PATH = "../data/RadChar-Baseline.h5"
```

```python
# open the file and print what sections are inside
with h5py.File(DATA_PATH, 'r') as f:
    for key in f.keys():
        print(key, '→ shape:', f[key].shape, '| dtype:', f[key].dtype)
```

```python
# load the raw I/Q signals and their labels
iq, labels = load_radchar(DATA_PATH)
print("IQ shape:", iq.shape)        # (1000000, 512) complex
print("Labels shape:", labels.shape)
```

```python
# look at one raw sample — IQ is stored as complex numbers (real=I, imaginary=Q)
sample = iq[0]
print('Length:', len(sample))
print('First 5 values:', sample[:5])
```

```python
# print all label fields and the label for the first sample
print('Label fields:', labels.dtype.names)
print('Sample 0 label:', labels[0])
```

```python
# how many samples of each signal type?
types = labels["signal_type"]
unique, counts = np.unique(types, return_counts=True)

for t, c in zip(unique, counts):
    print(f'Type {t}: {c} samples')
```
